```python

class ProductUrl(BaseModel):
    website: str = Field(..., description="The website of the product.")
    product_brand: str = Field(..., description="The brand of the product.")
    product_name: str = Field(..., description="The name of the product.")
    product_price: str = Field(..., description="The price of the product.")
    product_reviews_count_text: str = Field(..., description="The number of reviews for the product.")
    stars: str = Field(..., description="The star rating of the product.")
    url: str = Field(..., description="The URL of the product page.")
    


class ProductDetails(BaseModel):
    name: str = Field(..., description="The name of the product.")
    start_time: str = Field(..., description="The start time of the product mention in the video.")
    recommend: bool = Field(..., description="Whether the product is recommended or not. Only return yes or no.")
    reason: str = Field(..., description="Summary from the youtube video why this product is recommended or not.")
    details: str = Field(..., description="The details of the product.")
    product_url: ProductUrl = Field(..., description="The URL of the product page.")
    # price: str = Field(..., description="The price of the product.")
    # rating: str = Field(..., description="The rating of the product.")
    # review_count: str = Field(..., description="The number of reviews for the product.")
    # link: str = Field(..., description="The link to the product page.")

class YoutubeSummaryOutput(BaseModel):
    """
    A single, verifiable citation to a transcript snippet or video segment.
    Must correspond to a real snippet returned by the `search()` tool.
    """
    title: str = Field(..., description="The title of the YouTube video.")
    youtuber: str = Field(..., description="The name of the YouTuber.")
    url: str = Field(..., description="The URL of the YouTube video.")
    summary: str = Field(..., description="A brief summary of the video content.")
    category: str = Field(..., description="The category of the video content.")
    products: list[ProductDetails] = Field(..., description="A list of recommended products mentioned in the video.")
```

```python
# tool calling monitoring 

from pydantic_ai.messages import FunctionToolCallEvent

class NamedCallback:

    def __init__(self, agent):
        self.agent_name = agent.name

    async def print_function_calls(self, ctx, event):
        # Detect nested streams
        if hasattr(event, "__aiter__"):
            async for sub in event:
                await self.print_function_calls(ctx, sub)
            return

        if isinstance(event, FunctionToolCallEvent):
            tool_name = event.part.tool_name
            args = event.part.args
            print(f"TOOL CALL ({self.agent_name}): {tool_name}({args})")

    async def __call__(self, ctx, event):
        return await self.print_function_calls(ctx, event)
```

```python
# agent to summarize the youtube video
# recommend product if needed
# makeup advice and tips

youtube_instructions = """
You are a multilingual beauty content search and analysis agent specializing in YouTube videos.

You are given:
- A user query related to beauty content (e.g. "what foundation is she using?" or "what blush does she apply at 2:30?")
- Pre-searched video data including: title, summary, transcript (with timestamps), and possibly mentioned products.

Your goals:

1. **Query Understanding & Transcript Search**
   - Interpret the user query and find transcript segments that best answer it.
   - For each relevant segment, include its timestamp and text.
   - Focus only on relevant snippets — do not repeat the full transcript.
   - Use the provided summary to describe what the video is about (do not regenerate it).

2. **Product Extraction**
   - Identify all beauty products explicitly mentioned in the relevant transcript snippets.
   - If a product is implied but not named (e.g., user asks "what foundation?" but transcript says “I’m using this one from Sephora”), 
     infer likely product names from context or related information in the search results.
   - Each product should include:
        - `name`: the exact or inferred product name.
        - `details`: short info like purpose, texture, finish, or color tone.

3. **Video Context & Category**
   - Use the given video summary and metadata to identify the category:
        - makeup
        - skincare
        - haircare
   - Do not generate a new summary — just reuse the existing one.

4. **Output Format**
   Produce a structured JSON object following the `YoutubeSummaryOutput` schema:
"""

# beauty_instructions = """
# You are a multilingual beauty content search and analysis agent specializing in YouTube videos and beauty products.

# **Workflow to follow:**

# 1. **Query Understanding & Transcript Search**
#    - Take the user query and search the video transcript using `youtube_search_tool.search`.
#    - Find all transcript segments that are relevant to the query.
#    - Include the timestamp and the text of each relevant segment.
#    - Only return the segments that are relevant; do not repeat the entire transcript.

# 2. **Product Extraction from YouTube**
#    - Identify all products explicitly mentioned or highlighted in the relevant transcript segments.

# 3. **Sephora Product Search**
#    - For each product identified, search on Sephora using `sephora_search_tool.search_sephora`.

# 4. **Output**
#    - Return a structured JSON following the `YoutubeSummaryOutput` schema:
#         - Include all relevant product details from YouTube and Sephora.
#         - Include transcript segments relevant to the user query.
#         - Ensure all timestamps, product names, and details match the video content and query.
#    - Be factual and concise.

# """


beauty_instructions = """
You are a multilingual beauty content search and analysis agent specializing in YouTube videos and beauty products.

**Workflow to follow:**

1. **Query Understanding & Transcript Search**
   - Take the user query and search the video transcript using `youtube_search_tool.search`.
   - Find all transcript segments that are relevant to the query.
   - Include the timestamp and the text of each relevant segment.
   - Only return the segments that are relevant; do not repeat the entire transcript.

2. **Product Extraction from YouTube**
   - Identify all products explicitly mentioned or highlighted in the relevant transcript segments.
   - Summarize the reasone why the product is recommended or not based on the video content. 
   - Include all the details from the transcript that retrieved from the youtube_search_tool.search` 

3. **Sephora Product Search**
   - For each product identified, search on Sephora using `sephora_search_tool.search_sephora`.

4. **Output**
   - Return a structured JSON following the `YoutubeSummaryOutput` schema:
        - Include all relevant product details from YouTube and Sephora.
        - Include transcript segments relevant to the user query.
        - Ensure all timestamps, product names, and details match the video content and query.
   - Be factual and concise.

"""


youtube_agent = Agent(
    name='Youtube_Agent',
    instructions=beauty_instructions,
    tools=[youtuber_search_tool.search, sephora_search_tool.search_sephora],
    model='gpt-4o-mini',
    output_type=YoutubeSummaryOutput
)
callback = NamedCallback(youtube_agent)
```

```python
# question = "Any recommendation for NATASHA DENONA Hy-Glam?"
question = "Any recommendation foundation in the video?"
# question = "NATASHA DENONA Hy-Glam "
question = "Favorite foundation from Mindy"

result = await youtube_agent.run(
        user_prompt=question,
        event_stream_handler=callback,
    )
```

```python
def time_to_seconds(time_str: str) -> int:
    parts = time_str.split(':')
    parts = [int(p) for p in parts]
    
    if len(parts) == 2:  # MM:SS
        minutes, seconds = parts
        return minutes * 60 + seconds
    elif len(parts) == 3:  # HH:MM:SS
        hours, minutes, seconds = parts
        return hours * 3600 + minutes * 60 + seconds
    else:
        raise ValueError(f"Invalid time format: {time_str}")
```
